### 新手法「TAID」を用いた小規模日本語言語モデル「TinySwallow-1.5B」の公開
- 参考URL
1.   https://sakana.ai/taid-jp/
2.   https://github.com/SakanaAI/TinySwallow-ChatUI

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# 1. load model
device = "cuda" if torch.cuda.is_available() else "cpu"
repo_id = "SakanaAI/TinySwallow-1.5B-Instruct"
model = AutoModelForCausalLM.from_pretrained(repo_id)
tokenizer = AutoTokenizer.from_pretrained(repo_id)
model.to(device)
```

```python
print(device)
```

```python
# 2. prepare inputs
text = "生成ALについて簡単に教えてください。"
messages = [{"role": "user", "content": text}]
input_ids = tokenizer.apply_chat_template(messages, add_generation_prompt=True, return_tensors="pt")
```

```python
print(text)
```

```python
print(messages)
```

```python
print(input_ids)
```

```python
# 3. generate
output_ids = model.generate(
    input_ids.to(device),
    max_new_tokens=1024,
)
print(output_ids)
```

```python
print(output_ids[:, input_ids.shape[1] :])
```

```python
output_ids = output_ids[:, input_ids.shape[1] :]
generated_text = tokenizer.batch_decode(output_ids, skip_special_tokens=True)[0]
print(generated_text)
```
